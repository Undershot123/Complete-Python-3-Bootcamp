# Product Analysis

```python
import pandas as pd

date = {
    'Product_ID': range(101, 110),
    'Product_Name': ['Product_A', 'Product_B', 'Product_C', 'Product_D', 'Product_E',
                     'Product_F', 'Product_G', 'Product_H', 'Product_I'],
    'Price': [10.99, 12.49, 9.99, 15.99, 8.49, 14.99, 11.49, 13.99, 7.99],
    'Quantity_Sold': [150, 200, 250, 100, 300, 120, 180, 220, 400],
    'Category': ['Electronics','Home', 'Garden', 'Toys','Books', 'Clothing', 'Sports', 'Automotive', 'Health']
}

# Convert the dictionary to a pandas dataframe
df = pd.DataFrame(date)

df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Price'], df['Quantity_Sold'])
plt.show()
```
