```python
mystr = 'hello'
```

```python
#List Comprehension to create a list of characters
mylist = [] #empty list

for letter in mystr: #common method for making lists
    mylist.append(letter)
```

```python
mylist
```

```python
mylist = [letter for letter in mystr] #List comprehension method
```

```python
mylist
```

```python
mylist = [num for num in range(0,11)]
mylist
```

```python
#grab only even numbers
mylistt = [num for num in range(0,11) if num%2==0]
mylistt
```
