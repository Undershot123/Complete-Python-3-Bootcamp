```python
def player_guess():
    guess = ''
    while guess not in ['0', '1', '2']:
        guess = input("Pick a number: 0, 1 or 2")

    return int(guess) #Input always retruns a string, so we convert to int
```

```python
index = player_guess()
```

```python
def check_guess(mylist, guess):
    if mylist[guess] == 'O':
        print("Correct!")
    else:
        print("Wrong guess!")
        print(mylist)
        
```
