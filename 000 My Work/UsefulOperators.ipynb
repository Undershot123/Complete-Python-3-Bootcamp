# Range

```python
mylist = [1,2,3]
```

```python
for num in range(10):  # Range all the way up to 9
    print(num)
```

```python
for num in range(3,10):  # Start at 3 up to 9
    print(num)
```

```python
for num in range(0,11, 2):  # Start at 0 up to 10, evens only
    print(num)
```

```python
range(0,11) #Range is a generator object, it needs to be converted to a list to see all values
```

```python
list(range(0,11))
```

```python
index_count = 0

for letter in 'abcde':
    print('At index {} the letter is {}'.format(index_count, letter))
    index_count += 1
```

```python
word = 'abcde'
for item in enumerate(word): #returns tuples
    print(item)
```

```python
word = 'abcde'
for index,letter in enumerate(word): #returns tuples
    print(index)
    print(letter)
    print('\n')
```

```python
#Zip Function, combines two iterables into a list of tuples
mylist1 = [1,2,3]
mylist2 = ['a','b','c']
```

```python
zip(mylist1, mylist2) #zip is saved at the location below in the memory
```

```python
for item in zip(mylist1, mylist2): #Tuple unpacking
    print(item)
```

```python
list(zip(mylist1, mylist2)) # Alternative way to print the zip
```
